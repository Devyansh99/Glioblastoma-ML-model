## With JSON

```python
import os
import subprocess
import glob

# Paths to the two DICOM folders
dicom_folders = {
    "BRAINROUTINE-33106": r"D:\SEM_4\MLPR LAB (SEM-4)\Proj\02-06-2002-NA-BRAINROUTINE-33106 & 41468\02-06-2002-NA-BRAINROUTINE-33106\2.000000-t2Flairaxial ProcessedCaPTk-17693",
    "BRAINROUTINE-41468": r"D:\SEM_4\MLPR LAB (SEM-4)\Proj\02-06-2002-NA-BRAINROUTINE-33106 & 41468\02-06-2002-NA-BRAINROUTINE-41468\8.000000-ep2ddiffMDDWIPAT-81031"
}

# Path to dcm2niix.exe
dcm2niix_path = r"D:\Z\dcm2niix\dcm2niix.exe"

# Base output folder for NIfTI files
base_output_folder = r"D:\SEM_4\MLPR LAB (SEM-4)\Proj\Converted_NIfTI"
os.makedirs(base_output_folder, exist_ok=True)  # Ensure base output folder exists

# Convert each DICOM folder separately
for folder_name, dicom_folder in dicom_folders.items():
    output_folder = os.path.join(base_output_folder, folder_name)  # Create a subfolder for each DICOM folder
    os.makedirs(output_folder, exist_ok=True)  # Ensure output folder exists
    
    dicom_files = glob.glob(os.path.join(dicom_folder, "*"))  # Get all DICOM files
    
    for dicom_file in dicom_files:
        if os.path.isfile(dicom_file):  # Ensure it's a file
            filename = os.path.basename(dicom_file).split('.')[0]  # Extract filename
            output_name = os.path.join(output_folder, filename)  # Output filename

            print(f"Converting: {dicom_file} → {output_folder}")
            
            # Run dcm2niix for each file separately
            subprocess.run([dcm2niix_path, "-o", output_folder, "-f", filename, "-z", "y", dicom_file])

print("Conversion complete!")
```

## No JSON no ZIPPING

```python
import os
import subprocess
import glob

# Paths to the two DICOM folders
dicom_folders = {
    "BRAINROUTINE-33106": r"D:\SEM_4\MLPR LAB (SEM-4)\Proj\02-06-2002-NA-BRAINROUTINE-33106 & 41468\02-06-2002-NA-BRAINROUTINE-33106\2.000000-t2Flairaxial ProcessedCaPTk-17693",
    "BRAINROUTINE-41468": r"D:\SEM_4\MLPR LAB (SEM-4)\Proj\02-06-2002-NA-BRAINROUTINE-33106 & 41468\02-06-2002-NA-BRAINROUTINE-41468\8.000000-ep2ddiffMDDWIPAT-81031"
}

# Path to dcm2niix.exe
dcm2niix_path = r"D:\Z\dcm2niix\dcm2niix.exe"

# Base output folder for NIfTI files
base_output_folder = r"D:\SEM_4\MLPR LAB (SEM-4)\Proj\Converted_NIfTI"
os.makedirs(base_output_folder, exist_ok=True)  # Ensure base output folder exists

# Convert each DICOM folder separately
for folder_name, dicom_folder in dicom_folders.items():
    output_folder = os.path.join(base_output_folder, folder_name)  # Create a subfolder for each DICOM folder
    os.makedirs(output_folder, exist_ok=True)  # Ensure output folder exists
    
    dicom_files = glob.glob(os.path.join(dicom_folder, "*"))  # Get all DICOM files
    
    for dicom_file in dicom_files:
        if os.path.isfile(dicom_file):  # Ensure it's a file
            filename = os.path.basename(dicom_file).split('.')[0]  # Extract filename
            output_name = os.path.join(output_folder, filename)  # Output filename

            print(f"Converting: {dicom_file} → {output_folder}")
            
            # Run dcm2niix for each file separately (WITHOUT JSON and WITHOUT COMPRESSION)
            subprocess.run([dcm2niix_path, "-o", output_folder, "-f", filename, "-z", "n", "-b", "n", dicom_file])

print("Conversion complete!")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Function and gradient
def function(x):
    return x**2  # Quadratic function

def gradient(x):
    return 2*x  # Derivative

# Adam Parameters
eta_0 = 0.33     # Initial learning rate
beta1 = 0.9      # Momentum decay
beta2 = 0.999    # RMSprop decay
epsilon = 1e-8   # Prevent division by zero
weight_decay = 0.01  # Weight decay (AdamW)
decay_rate = 0.05  # Exponential decay rate

# Initialize
x = 5.0
m, v = 0, 0
iterations = 50
x_values = []
lr_values = []

# Adam with Learning Rate Decay
for t in range(1, iterations + 1):
    g = gradient(x)

    # Compute learning rate with exponential decay
    eta_t = eta_0 * np.exp(-decay_rate * t)
    lr_values.append(eta_t)

    # Adam update steps
    m = beta1 * m + (1 - beta1) * g
    v = beta2 * v + (1 - beta2) * g**2
    m_hat = m / (1 - beta1**t)
    v_hat = v / (1 - beta2**t)
    
    # Weight decay (AdamW)
    x -= (eta_t / (np.sqrt(v_hat) + epsilon)) * m_hat + weight_decay * x
    x_values.append(x)

# Plot function and optimization path
x_plot = np.linspace(-6, 6, 100)
y_plot = function(x_plot)

plt.figure(figsize=(8, 5))
plt.plot(x_plot, y_plot, label="f(x) = x^2")
plt.scatter(x_values, [function(xi) for xi in x_values], color='red', label="Adam Steps")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Adam with Learning Rate Decay")
plt.legend()
plt.grid()
plt.show()

# Plot Learning Rate Decay over time
plt.figure(figsize=(8, 4))
plt.plot(range(1, iterations + 1), lr_values, label="Learning Rate Decay")
plt.xlabel("Iterations")
plt.ylabel("Learning Rate")
plt.title("Exponential Learning Rate Decay")
plt.legend()
plt.grid()
plt.show()
```
